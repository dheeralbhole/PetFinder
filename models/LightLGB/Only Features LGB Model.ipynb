```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import lightgbm as lgb

from sklearn.model_selection import train_test_split
```

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
df_train.head()
```

```python

X_train = df_train.drop(["Id", "Pawpularity"], axis=1)
y_train = df_train["Pawpularity"] / 100
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25)
train_data = lgb.Dataset(X_train, y_train)
val_data = lgb.Dataset(X_val, y_val)
```

```python
params = {
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'learning_rate': 0.05,
    'num_round': 100,
    'verbosity': -1,
}

classifier = lgb.train(params=params,
                       train_set=train_data,
                       valid_sets=[train_data, val_data],
                       valid_names=["train", "valid"],
                       verbose_eval=1)
```

```python
#Now submit to the competition using the model:
test_df = pd.read_csv('test.csv')
x_test_submission = test_df.drop(['Id'],axis=1) 
test_df['Pawpularity'] = classifier.predict(x_test_submission) 
submission_df = test_df[['Id','Pawpularity']]
submission_df.to_csv("submission.csv", index=False)
submission_df.head()
```
